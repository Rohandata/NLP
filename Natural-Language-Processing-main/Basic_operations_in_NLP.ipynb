##Tokenization  


Tokenization is a way of separating a piece of text into smaller units called tokens. Here, tokens can be either words, characters, or subwords. Hence, tokenization can be broadly classified into 3 types â€“ word, character, and subword (n-gram characters) tokenization.



```python
x="I love to learn natural langage processing and deep dive into it."
type(x)
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
z=word_tokenize(x)
y=sent_tokenize(x)
print(z)
print(y)
```

#STOPWORDS

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```python
print(tokens)
filtered_sen=[]
for w in tokens:
  if w not in stop_words:
    filtered_sen.append(w)
print(filtered_sen)
```

```python
filtered_sen=[w for w in tokens if not w in stop_words]
filtered_sen
```

#POS TAGGING

```python
nltk.download('state_union')
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
train_txt=state_union.raw("/content/pg11.txt")
sample_txt=state_union.raw("/content/pg11.txt")
custom_Sent=PunktSentenceTokenizer(train_txt)
sample_sent=custom_Sent.tokenize(sample_txt)
def process_content():
  try:
    for i in sample_sent:
      words=nltk.word_tokenize(i)
      tagged=nltk.pos_tag(words)
      print(tagged)
  except Exception as e:
    print(str(e))
process_content()
```
