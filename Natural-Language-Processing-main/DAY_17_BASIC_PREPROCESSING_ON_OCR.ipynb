```python
image_path_in_colab="/content/Testproc1.jpg"
extract = pytesseract.image_to_string(Image.open(image_path_in_colab))
print(extract)
#type(extract)
```

#BASIC PREPROCESSING USING NLTK

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
print(word_tokenize(extract))
```

```python
from nltk.tokenize import sent_tokenize
print(sent_tokenize(extract))
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords 
   
stop_words = set(stopwords.words('english')) 
words = extract.split() 
text_without_sw = [token for token in words if not token in stop_words]
print(text_without_sw)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
from nltk.tag import pos_tag
sent = nltk.word_tokenize(extract)
sent = nltk.pos_tag(sent)
print(sent)
```

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
from nltk import ne_chunk, pos_tag, word_tokenize
print(nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(extract))))
```
